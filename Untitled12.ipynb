```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#membaca image
img = cv2.imread('/content/buah.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Ukuran image
rows, cols, ch = img.shape

#menentukan jarak transmisi
tx,ty = 50,30

M = np.float32((
    [1,0,tx],
    [0,1,ty]
))

img_translated = cv2.warpAffine(img, M, (cols, rows))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img_translated)
plt.title('Translated Image')
plt.show()
```

```python
#menentukan titik pusat
center = (cols/2, rows/2)
#tentukan sudut rotasi
angle = 45 #postif => berlawanan awach jarum jam
#skala tetap
scale = 1.0 #default

#membuat matriks rotasi
M = cv2.getRotationMatrix2D(center, angle, scale)
#menerapkan rotasi pada citra
img_rotated = cv2.warpAffine(img, M, (cols,rows))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img_rotated)
plt.title('Rotated Image')
plt.show
```

```python
sx, sy = 2, 3
M = np.float32((
    [sx, 1,0],
    [0,sy,1]
))
img_scaled = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(122)
plt.imshow(img_scaled)
plt.title('Scaled Image')
```

```python
x1, y1 = 100, 50
x2, y2 = 400, 300

img_cropped = img[y1:y2, x1:x2]

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(122)
plt.imshow(img_scaled)
plt.title('Scaled Image')
plt.show()

```

```python
# Parameter ripple
amplitudo_x = 10
amplitudo_y = 10
wavelength_x = 30
wavelength_y = 30

# grid koordinat piksel
x, y = np.meshgrid(np.arange(cols), np.arange(rows))

# menghitung posisi baru (x,y) setiap piksel
x_new = x + amplitudo_x * np.sin(2 * np.pi * y / wavelength_x)
y_new = y + amplitudo_y * np.sin(2 * np.pi * x / wavelength_y)

# normalisasi agar sesuai dengan ukuran gambar
x_new = np.clip(x_new, 0, cols-1).astype(np.float32)
y_new = np.clip(y_new, 0, rows-1).astype(np.float32)

# menerapkan transformasi ripple
img_warped = cv2.remap(img, x_new, y_new, cv2.INTER_LINEAR)

# Menampilkan gambar
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img_warped)
plt.title('Warped Image (Ripple Effect)')
plt.show()
```

```python
center_x, center_y = cols/2, rows/2

strength = 5.0

radius = min(center_x, center_y) * 0.8

x, y = np.meshgrid(np.arange(cols), np.arange(rows))

x_rel = x - center_x
y_rel = y - center_y

r = np.sqrt(x_rel**2 + y_rel**2) # Corrected: use **2 for squaring

theta = np.arctan2(y_rel, x_rel)

theta_new = theta + strength * (1 - (r/radius))

theta_new[r > radius] = theta[r > radius]

x_new = center_x + r * np.cos(theta_new)
y_new = center_y + r * np.sin(theta_new)

x_new = np.clip(x_new, 0, cols-1).astype(np.float32)
y_new = np.clip(y_new, 0, rows-1).astype(np.float32)

img_twirl = cv2.remap(img, x_new, y_new, cv2.INTER_LINEAR)

# Menampilkan gambar
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img_twirl)
plt.title('Twirled Image')
plt.show()
```
