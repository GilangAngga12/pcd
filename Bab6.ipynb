```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/sample_data/gambar bab5.jpeg', cv2.IMREAD_GRAYSCALE)

kernel_size = 3

# Create a kernel using numpy.ones
kernel = np.ones((kernel_size, kernel_size), np.uint8)
min_filter = cv2.erode(img, kernel, iterations=1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off') # Turn off axis for cleaner visualization

plt.subplot(2, 2, 2)
plt.imshow(min_filter, cmap='gray')
plt.title('Min Filtered Image (Erosion)')
plt.axis('off') # Turn off axis for cleaner visualization

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_min_filter = cv2.calcHist([min_filter], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_min_filter)
plt.title('Histogram of Min Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
max_filter = cv2.dilate(img, kernel)

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(max_filter, cmap='gray')
plt.title('Max Filtered Image (Dilation)')
plt.axis('off')

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_max_filter = cv2.calcHist([max_filter], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_max_filter)
plt.title('Histogram of Max Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
mean_filter = cv2.blur(img, (kernel_size, kernel_size))

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(mean_filter, cmap='gray')
plt.title('Mean Filtered Image')
plt.axis('off')

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_mean_filter = cv2.calcHist([mean_filter], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_mean_filter)
plt.title('Histogram of mean Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
median_filter = cv2.medianBlur(img, kernel_size)

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(median_filter, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_median_filter = cv2.calcHist([median_filter], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_median_filter)
plt.title('Histogram of median Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**#LPF**

```python
sigma = 1.0
gaussian_filter = cv2.GaussianBlur(img, (kernel_size, kernel_size), sigma)

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(gaussian_filter, cmap='gray')
plt.title('Gaussian Filtered Image')
plt.axis('off')

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_gaussian_filter = cv2.calcHist([gaussian_filter], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_gaussian_filter)
plt.title('Histogram of gaussian Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')


plt.show()
```

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel Filtered Image')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X Filtered Image')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
lapcian = cv2.Laplacian(img, cv2.CV_64F)

# Convert the Laplacian image to 8-bit unsigned integer format
lapcian_8bit = cv2.convertScaleAbs(lapcian)

plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(lapcian, cmap='gray')
plt.title('Laplacian Filtered Image')
plt.axis('off')

# Calculate and plot the histogram of the original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_original)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Calculate and plot the histogram of the min-filtered image
hist_laplacian_filter = cv2.calcHist([lapcian_8bit], [0], None, [256], [0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_laplacian_filter)
plt.title('Histogram of laplacian Filtered Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

print(sobel_x)

print(sobel_y)

print(lapcian)
plt.tight_layout()
plt.show()
```

```python
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
sobel_uint8 = cv2.convertScaleAbs(sobel_combined)
lapcian_8bit = cv2.convertScaleAbs(lapcian)

plt.figure(figsize=(12, 8))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Combined Filtered Image')
plt.axis('off')

print('sobel', sobel_combined)
print('laplacian', lapcian_8bit)

plt.tight_layout()
plt.show()
```
