```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/kucing3.jpeg', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([img],[0],None,[256],[0,256])

plt.figure(figsize=(12,5))

plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,1)
plt.hist(img.ravel(),256,[0,256])
# plt.plot(hist,color = 'black')
plt.bar(range(256), hist.flatten(), color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('jumlah piksel')

max_idx = np.argmax(hist)
plt.axvline(x=max_idx, color='red', linestyle='--', label=f'Puncak = {max_idx}')
plt.legend()

# plt.show()
```

```python
equalized = cv2.equalizeHist(img)
plt.figure(figsize=(12,8)) # Increased figure height

#original image
plt.subplot(2,2,1)
plt.title('Equalized Image')
plt.imshow(equalized, cmap='gray')

#equalized imagge
plt.subplot(2,2,2)
plt.title('equalized image')
plt.imshow(equalized, cmap='gray')

#histogram original image
plt.subplot(2,2,3)
plt.hist(equalized.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')
plt.title('Histogram asli')

#histogram equalized image
plt.subplot(2,2,4)
plt.hist(equalized.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')
plt.title('Histogram equalized')

plt.tight_layout() # Add tight_layout to prevent overlap
plt.show()
```

```python
_, tresh = cv2.threshold(equalized, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10,5))

#original image
plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(equalized, cmap='gray')

#threshold image
plt.subplot(2,2,2)
plt.title('Threshold Image')
plt.imshow(tresh, cmap='gray')

plt.subplot(2,2,3)
plt.hist(img.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')
plt.title('Histogram asli')

plt.subplot(2,2,4)
plt.title('Histogram tresholding')
plt.hist(img.ravel(),2,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')
plt.title('Histogram equalized')

plt.tight_layout()
plt.show()
```

```python
bright = cv2.convertScaleAbs(img, alpha=1, beta=100)
dark = cv2.convertScaleAbs(img, alpha=1, beta=-50)
# alpha = kontras batasnya 1, jika lebih dari 1 kontras naik jika kurang kontras turun
#beta = kecerahan batasnya 0 lebih dari 0 kecerahan meningkat jika kurang dari 0 keceragan menurun
plt.figure(figsize=(12,4))

plt.subplot(2,3,1)
plt.title('gambar gelap')
plt.imshow(dark, cmap='gray')

plt.subplot(2,3,2)
plt.title('gambar terang')
plt.imshow(bright, cmap='gray')

plt.subplot(2,3,3)
plt.title('gambar asli')
plt.imshow(img, cmap='gray')

plt.subplot(2,3,4)
plt.title('histogram gelap')
plt.hist(dark.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')

plt.subplot(2,3,5)
plt.title('histogram terang')
plt.hist(bright.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')

plt.subplot(2,3,6)
plt.title('histogram asli')
plt.hist(img.ravel(),256,[0,256], color = 'black')


plt.tight_layout()
plt.show()
```

```python
inverted = 255 - img #invers (citra negatif)

plt.subplot(2,2,1)
plt.title('gambar asli')
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title('gambar negatif')
plt.imshow(inverted, cmap='gray')

plt.subplot(2,2,3)
plt.title('histogram terang')
plt.hist(img.ravel(),256,[0,256], color = 'black')
plt.xlabel('tingkat keabuan (0-255)')
plt.ylabel('frekuensi piksel')

plt.subplot(2,2,4)
plt.title('histogram negatif')
plt.hist(inverted.ravel(),256,[0,256], color = 'black')

plt.tight_layout()
plt.show()
```

```python
blur = cv2.blur(img, (5,5))

edges = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)#ksize wajib ganjil

plt.subplot(2,3,1)
plt.title('gambar asli')
plt.imshow(img, cmap='gray')

plt.subplot(2,3,2)
plt.title('gambar blur')
plt.imshow(blur, cmap='gray')

plt.subplot(2,3,3)
plt.title('gambar edge')
plt.imshow(edges, cmap='gray')

plt.subplot(2,3,4)
plt.title('histogram asli')
plt.hist(img.ravel(),256,[0,256], color = 'black')

plt.subplot(2,3,5)
plt.title('histogram blur')
plt.hist(blur.ravel(),256,[0,256], color = 'black')

plt.subplot(2,3,6)
plt.title('histogram edge')
plt.hist(edges.ravel(),256,[0,256], color = 'black')

plt.tight_layout()

plt.show()
```

```python
norm = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

plt.figure(figsize=(12,4))

plt.subplot(2,2,1)
plt.title('gambar asli')
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title('gambar normalisasi')
plt.imshow(norm, cmap='gray')

plt.subplot(2,2,3)
plt.title('histogram asli')

plt.hist(img.ravel(),256,[0,256], color = 'black')

plt.subplot(2,2,4)
plt.title('histogram normalisasi')
plt.hist(norm.ravel(),256,[0,256], color = 'black')

plt.tight_layout()

plt.show()
```
