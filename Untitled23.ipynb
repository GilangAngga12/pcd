```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#membaca image
img = cv2.imread('/content/sample_data/images.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#ukuran image
rows, cols, _ = img.shape

#menentukan jarak transisi
tx,ty = 50,30

M = np.float32((
    [1, 0, tx],
    [0, 1, ty]
))

img_translated = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_translated)
plt.title('Translated')
plt.show()
```

```python
#menentukan titik pusat
center = (cols/2, rows/2)

# trntukan sudut rotasi
angle = 45 #positif=> memutar berlawanan arah jarum jam

# skala tetap
scale = 1.0

M = cv2.getRotationMatrix2D(center, angle, scale)
img_rotated = cv2.warpAffine(img, M, (cols, rows))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_rotated)
plt.title('Rotated')
```

```python
sx, sy = 2, 3
M = np.float32((
    [sx, 1, 0 ],
    [0, sy, 1]
))

img_scaled = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_scaled)
plt.title('Scaled')
```

```python
x1, y1 = 50, 50
x2, y2 = 400, 300

img_cropped = img[y1:y2, x1:x2]

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_cropped)
plt.title('Cropped')
```

```python
shx = 0.4
shy = 0.0

M = np.float32((
    [1, shx, 0 ],
    [shy, 0, 1]
))

img_sheared = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_sheared)
plt.title('Sheared')
```

```python
# parameter ripple
amplitudo_x = 10
amplitudo_y = 10
wavelength_x = 20
wavelength_y = 20

# grid koordinat pixel
x, y = np.meshgrid(np.arange(cols), np.arange(rows))


phase_x = amplitudo_x * np.sin(2 * np.pi * x / wavelength_x)
phase_y = amplitudo_y * np.sin(2 * np.pi * y / wavelength_y)

# menghitung posisi baru
x_new = x + amplitudo_x * np.sin(2 * np.pi * x / wavelength_x)
y_new = y + amplitudo_y * np.sin(2 * np.pi * y / wavelength_y)

# normalisasi agar sesuai dengan ukuran gambar
x_new = np.clip(x_new, 0, cols - 1).astype(np.float32)
y_new = np.clip(y_new, 0, rows - 1).astype(np.float32)

# menerapkan transformasi ripple
img_warped = cv2.remap(img, x_new, y_new, cv2.INTER_LINEAR)


plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(img_warped)
plt.title('Ripple')
```
