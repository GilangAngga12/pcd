```python
# Soal No 1
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2 as cv

gambar1 = mpimg.imread("/content/sample_data/frekuensi tinggi.jpg")
gambar2 = mpimg.imread("/content/sample_data/images.jpeg")


plt.figure(figsize=(10,5))

plt.subplot(2,3,1)
plt.title("Gambar cerah merata")
plt.imshow(gambar1, cmap="gray")

plt.subplot(2,3,2)
plt.title("Gambar Cerah")
plt.imshow(gambar2, cmap="gray")



plt.subplot(2,3,4)
plt.title("histogram gambar cerah merata")
plt.hist(gambar1.ravel(), 256, [0,256], color="yellow")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")

plt.subplot(2,3,5)
plt.title("histogram gambar terang")
plt.hist(gambar2.ravel(), 256, [0,256], color="yellow")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")



plt.show()
```

```python
equalized1 = cv.equalizeHist(cv.cvtColor(gambar1, cv.COLOR_BGR2GRAY))
equalized2 = cv.equalizeHist(cv.cvtColor(gambar2, cv.COLOR_BGR2GRAY))

plt.figure(figsize=(10,5))

#Gambar equalized
plt.subplot(2, 3, 1)
plt.title("Gambar Equalized 1")
plt.imshow(equalized1, cmap="gray")

plt.subplot(2, 3, 2)
plt.title("Gambar Equalized 2")
plt.imshow(equalized2, cmap="gray")


#gambar histogram asli
plt.subplot(2,3,4)
plt.hist(equalized1.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")
plt.title("Histogram Equalized 1")

plt.subplot(2,3,5)
plt.hist(equalized2.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")
plt.title("Histogram Equalized 2")




plt.show()
```

```python
from scipy import datasets, ndimage
from matplotlib import pyplot as plt

face = datasets.face()
blurred_face = ndimage.gaussian_filter(face, sigma=3)

fig, ax = plt.subplots(ncols=2, figsize=(16, 8))
ax[0].imshow(face)
ax[0].set_title('Gambar asli')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(blurred_face)
ax[1].set_title('Gambar blur')
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.show()
```
