```python
# Soal No 1
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2 as cv

gambar1 = mpimg.imread("/content/gambar cerah merata tugas.jpg")
gambar2 = mpimg.imread("/content/gambar terlalu terang.jpg")
gambar3 = mpimg.imread("/content/gamabar gelap.jpg")


plt.figure(figsize=(10,5))

plt.subplot(2,3,1)
plt.title("Gambar cerah merata")
plt.imshow(gambar1, cmap="gray")

plt.subplot(2,3,2)
plt.title("Gambar Cerah")
plt.imshow(gambar2, cmap="gray")

plt.subplot(2,3,3)
plt.title("Gambar Gelap")
plt.imshow(gambar3, cmap="gray")

plt.subplot(2,3,4)
plt.title("histogram gambar cerah merata")
plt.hist(gambar1.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")

plt.subplot(2,3,5)
plt.title("histogram gambar terang")
plt.hist(gambar2.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")

plt.subplot(2,3,6)
plt.title("histogram gambar gelap")
plt.hist(gambar3.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")

plt.show()
```

```python
equalized1 = cv.equalizeHist(cv.cvtColor(gambar1, cv.COLOR_BGR2GRAY))
equalized2 = cv.equalizeHist(cv.cvtColor(gambar2, cv.COLOR_BGR2GRAY))
equalized3 = cv.equalizeHist(cv.cvtColor(gambar3, cv.COLOR_BGR2GRAY))

plt.figure(figsize=(10,5))

#Gambar equalized
plt.subplot(2, 3, 1)
plt.title("Gambar Equalized 1")
plt.imshow(equalized1, cmap="gray")

plt.subplot(2, 3, 2)
plt.title("Gambar Equalized 2")
plt.imshow(equalized2, cmap="gray")

plt.subplot(2, 3, 3)
plt.title("Gambar Equalized 3")
plt.imshow(equalized3, cmap="gray")

#gambar histogram asli
plt.subplot(2,3,4)
plt.hist(equalized1.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")
plt.title("Histogram Equalized 1")

plt.subplot(2,3,5)
plt.hist(equalized2.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")
plt.title("Histogram Equalized 2")

plt.subplot(2,3,6)
plt.hist(equalized3.ravel(), 256, [0,256], color="red")
plt.xlabel("Tingkat keabuan [0-255]")
plt.ylabel("frekuensi pixel")
plt.title("Histogram Equalized 3")


plt.show()
```
