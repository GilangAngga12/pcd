```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from skimage.morphology import (
    square,
    diamond,
    rectangle,
    disk,
    cube,
    octahedron,
    ball,
    octagon,
    star
)
#Generate 2D and 3D structuring elements.
struc_2d ={
  "square(15)": square(15),
  "rectangle(15, 10)": rectangle(15,10),
  "diamond(7)": diamond(7),
  "disk(7)": disk(7),
  "octagon (7, 4)": octagon (7, 4),
  "star(5)": star(5)
  }

struc_3d = {
    "cube(11)": cube (11),
    "octahedron (5)": octahedron (5),
    "ball(5)": ball(5)
}

#visualize the elements.
fig = plt.figure(figsize = (8, 8))

idx = 1
for title, struc in struc_2d.items():
  ax = fig.add_subplot(3, 3, idx)
  ax.imshow(struc, cmap="Paired", vmin=0, vmax=12)
  for i in range(struc.shape[0]):
    for j in range(struc.shape[1]):
      ax.text(j, i, struc[i, j],ha="center", va="center", color="w")
    ax.set_axis_off()
ax.set_axis_off()
ax.set_title(title)
idx += 1

for title, struc in struc_3d.items():
  ax = fig.add_subplot(3, 3, idx, projection=Axes3D.name)
  ax.voxels(struc)
  ax.set_title(title)
  idx += 1

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.data import coins
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.morphology import label, closing, square
from skimage.measure import regionprops
from skimage.color import lab2rgb
%matplotlib inline

def show(img, cmap=None):
  cmap = cmap or plt.cm.gray
  fig, ax = plt.subplots(1, 1, figsize=(8, 6))
  ax.imshow(img, cmap=cmap)
  ax.set_axis_off()
  plt.show()

img = coins()
show(img)
```

```python
threshold_otsu(img)
show(img > 107)
```

```python
img_bin = clear_border(
    closing(img > 120,square(5)))
show(img_bin)
```

```python
labels = label(img_bin)
show(labels, cmap=plt.cm.rainbow)
```

```python
regions = regionprops(labels)

boxes = np.array([label['BoundingBox']
                  for label in regions if label['Area'] > 100])

print(f"There are{len(boxes)} coins.")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.imshow(img, cmap=plt.cm.gray)

ax.set_axis_off()

#Get the coordinates of the baxes.
xs = boxes[:, [1, 3]].mean(axis=1)
ys = boxes [:, [0, 2]].mean(axis=1)

#we reorder the boxes by increasing

#column first, and row second.

for row in range(4):
  #we select the coins in each of the four rows.
  if row < 3:
    ind = ((ys[6 * row] <= ys) & (ys<ys[6*row + 6]))
  else:
    ind = (ys[6*row] <= ys)

#we reorder by increasing x coordinate.

  ind = np.nonzero(ind) [0]
  reordered = ind[np.argsort(xs[ind])]
  xs_row = xs[reordered]
  ys_row = ys[reordered]

#we display the coin number.
  for col in range(6):
    n = 6 * row + col
    ax.text(
    xs_row[col] - 5, ys_row[col] +5, str(n), fontsize=20)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img = cv2.imread('/content/sample_data/500-Rupiah-T-B-Simatupang.jpg',0)
_, binary = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

kernel = np.ones((5,5),np.uint8)
eroded = cv2.erode(binary,kernel)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.title("grayscale")
plt.imshow(img,cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("binary")
plt.imshow(binary,cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("eroded")
plt.imshow(eroded,cmap='gray')
plt.axis('off')

plt.show()
```

```python
dilated = cv2.dilate(binary,kernel)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title("grayscale")
plt.imshow(img,cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("binary")
plt.imshow(binary,cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Dilated")
plt.imshow(eroded,cmap='gray')
plt.axis('off')

plt.show()
```

```python
#OPENING
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
opening = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
opening = cv2.morphologyEx(opening, cv2.MORPH_GRADIENT, kernel)
opening = cv2.morphologyEx(opening, cv2.MORPH_TOPHAT, kernel)
opening = cv2.morphologyEx(opening, cv2.MORPH_BLACKHAT, kernel)


plt.subplot(2,3,1)
plt.title("binary")
plt.imshow(binary,cmap='gray')
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Opening")
plt.imshow(eroded,cmap='gray')
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Closing")
plt.imshow(opening,cmap='gray')
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Gradient")
plt.imshow(eroded,cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Tophat")
plt.imshow(opening,cmap='gray')
plt.axis('off')

plt.subplot(2,3,6)
plt.title("Blackhat")
plt.imshow(opening,cmap='gray')
plt.axis('off')

plt.show()
```
